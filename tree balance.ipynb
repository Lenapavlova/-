```python
class Node:
    def __init__(self, data) -> None:
        self.data =data
        self.left = None
        self.right = None
        self.reset_height = 1
```

```python
class Tree:
    def __init__(self):
        self.root = None

    def rec_append(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self.append_(self.root, data)
            if abs(self.root.bf()) == 2:
                    self.root = self.balance(self.root)

    def append_(self, cursor, data):
         if cursor.data > data:
            if cursor.left is None:
                cursor.left = Node(data)
                added = True
            else:
                self.append_(cursor.left, data)
                if abs(cursor.bf()) == 2:
                    cursor.left = self.balance(cursor.left)
         else:
            if cursor.right is None:
                cursor.right = Node(data)
                added = True
            else:
                self.append_(cursor.right, data)
                if abs(cursor.bf()) == 2:
                    cursor.right = self.balance(cursor.right)
         cursor.reset_height()

    def balance(self, cursor):
        if cursor.bf() > 0:
            if cursor.right.bf() < 0:
                return big_left_turn(cursor)
            else:
                return left_turn(cursor)
        else:
            if cursor.right.bf() > 0:
                return big_right_turn(cursor)
            else:
                return right_turn(cursor)



    def print_(self, cursor, depth):
        if cursor is not None:
            print(depth * '\t', cursor.data)
            print(depth * '\t', 'left:')
            self.print_(cursor.left, depth + 1)
            print(depth * '\t', 'right:')
            self.print_(cursor.right, depth + 1)

    def get_depths(self):
        left_depth = self._get_depth(self.root.left)
        right_depth = self._get_depth(self.root.right)
        return left_depth, right_depth

    def _get_depth(self, cursor):
        if cursor is None:
            return 0
        else:
            left_depth = self._get_depth(cursor.left)
            right_depth = self._get_depth(cursor.right)
            return max(left_depth, right_depth) + 1

    def lheight(self):
        return self.left.height if self.left is not None else 0

    def rheight(self):
        return self.right.height if self.right is not None else 0

    def reset_height(self):
        self.height = max(self.lheight(), self.rheight()) + 1

    def bf(self):
        return self.rheight() - self.lheight()

    def left_turn(q):
        p = q.right
        q.right = p.left
        p.left = q
        q.reset_height()
        p.reset_height()
        return p

    def right_turn(p):
        q = p.left
        p.left = q.right
        q.right = p
        p.reset_height()
        q.reset_height()
        return p

    def big_left_turn(q):
        p = q.right
        s = p.left
        right_turn(p)
        left_turn(q)
        s.reset_height()
        return s


    def big_right_turn(p):
        q = p.left
        s = q.right
        left_turn(q)
        right_turn(p)
        s.reset_height()
        return s

```
